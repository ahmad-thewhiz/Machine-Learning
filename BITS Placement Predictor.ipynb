```python
#import NumPy, Pandas and Matplot

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load dataset

df = pd.read_csv("bits.csv") 
```

```python
#print dataset

df 
```

```python
m = len(df['year']) #find the total number of available objetcs

print ('Number of training examples (m): ',m)
```

```python
#plot graph of available labeled datasets

plt.scatter(df['year'], df['avg'], marker='x', c='b') #plot the points on the graph using matplot

plt.title("BITS Pilani year wise placements")

plt.xlabel("Year")

plt.ylabel("Salary in LPA")
```

```python
#import linear refression model from SKLearn and create it's object

from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.avg)
```

```python
#regression coefficient 'm'

reg.coef_
```

```python
#regression intercept 'b'

reg.intercept_
```

```python
#plotting of best-fit linear regression model for the given dataset

plt.plot(df.year,reg.predict(df[['year']]),color='green')
```

```python
#plotting of the graph with LR model line and the scattered data points

plt.scatter(df['year'], df['avg'], marker='x', c='b') #plot the points on the graph using matplot
plt.title("BITS Pilani year wise placements")
plt.xlabel("Year")
plt.ylabel("Salary in LPA")
plt.plot(df.year,reg.predict(df[['year']]),color='green')
```

```python
#predicting y for a unique value of x

reg.predict(np.array([2023]).reshape(1, 1))
```
