### What is Matplotlib?

* Matplotlib is a popular Python library used for creating visualizations and plots. It provides a way to display data in a graphical format, making it easier to understand and analyze.

1. Plotting Library: Matplotlib is a plotting library that allows you to create various types of plots, charts, and graphs.

2. Wide Range of Plot Types: Matplotlib supports a wide range of plot types, including line plots, scatter plots, bar plots, histogram plots, pie charts, and many more.

3. Customization Options: It offers extensive customization options to control the appearance of plots. You can customize aspects such as colors, line styles, markers, labels, titles, axes, grids, and legends.

4. Publication-Quality Plots: Matplotlib enables you to create high-quality plots suitable for publication or presentations. You can save the plots in various file formats, such as PNG, JPEG, PDF, or SVG.

* Documentation Link : https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.plot.html

# A1: Line Plot & Scatter

```python
#plotting points on the graph

x = [1,2,3,4,5]
y = [1,8,27,64,125]

plt.scatter(x, y)
plt.show()
```

```python
#plotting a basic linear graph using custom values

x = [1,2,3,4,5]
y = [1,8,27,64,125]

plt.plot(x,y)

plt.title('First Graph', fontdict={'fontname': 'serif', 'fontsize': 20}) #Title with font-styling

plt.xlabel('x=x') #Label on x-axis
plt.ylabel('y=x^3') #label on y-axis

plt.show()
```

```python
#smooth curve

x=np.arange(0,5,0.1)
y=x*x*x

plt.plot(x,y)
plt.show
```

```python
#plotting points and curve together

x = [1,2,3,4,5]
y = [1,8,27,64,125]

plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
#plotting basic log function on graph

x = np.arange(1, 50, 0.1) #x array has values ranging from 1 to 50 with the intervals of 0.1
y = np.log(x) #y array has the log(x) values
plt.plot(x, y)
```

```python
#changing scale of axes

x = [1,2,3,4,5]
y = [1,8,27,64,125]

plt.xticks([0,1,2,3,3.5,4,5])
plt.yticks([0,5,10,15,20,45])

plt.plot(x,y)
plt.show()
```

```python
#adding legend to the graph

plt.plot(x,y, label='x^3', c='g') # c='g'  changes the colour to green
plt.legend()
plt.show()
```

```python
#some more modifications

plt.plot(x,y,c='r',linewidth=2,marker='.', linestyle="--", markersize=5, markeredgecolor='blue')
#plt.plot(x,y,"g--") will do the same job

plt.show() 
```

# A2: Bubble Chart

```python
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 300, 370]
colors = ['black', 'red', 'violet', 'blue', 'yellow', 'green'] #array of differnet colors

plt.scatter(year, salary, s = population, c = colors) #s=population will cause each data point to have size with respect to the value stored in population at that particular index.
#year[0] vs salary[0] with size=population[0] and color=colors[0]
plt.title('Bubble Chart')
plt.show()
```

```python
plt.scatter(year, salary, s = population, c = 'b', alpha = 0.2)
#alpha function is used to set transperency of the data points, lower value of alpha represents more % of transperency
plt.title('Bubble Chart')
plt.show()
```

```python
c = np.arange(12,24,2)

plt.scatter(year, salary, s=population, marker='^', c=c, alpha=0.4, edgecolor='blue')
plt.show()
```

```python
c = np.arange(12,24,2)

plt.scatter(year, salary, s=population, marker='^', c=c, alpha=0.4, edgecolor='blue')

plt.text(year[0]+0.2, salary[0]+0.2, population[0])
plt.text(year[5]-0.2, salary[5]-0.7, population[5])
#plt.text(x axis +/- a, y axis +/- b, value) a & b are used to change the location of text

plt.show()
```

# A3: Pie Chart

```python
labels = ["A", "B", "C", "D", "E"]
sizes = [6, 2, 5, 3, 1] 
colors = ['green', 'purple', 'red', 'pink', 'blue']
explode = [0.4, 0.1, 0.2, 0, 0.3]

plt.pie(sizes, colors = colors, labels = labels, autopct='%d')

plt.title("Split among classes")
plt.axis('equal')
#plt.axis('equal') - ensures that the pie chart is displayed with a proper aspect ratio
plt.show()
```

```python
labels = ["A", "B", "C", "D", "E"]
sizes = [6, 2, 5, 3, 1] 
colors = ['green', 'purple', 'red', 'pink', 'blue']
explode = [0.3, 0.1, 0.2, 0, 0.2]

plt.pie(sizes, colors = colors, explode = explode, labels = labels,  autopct = '%.1f', startangle = 100, counterclock = False)

#explode - the extent to which a particular slice of the pie chart is separated or "exploded" from the center
#autopct - controls the formatting of the percentage values displayed on each pie slice
#startangle -  the angle at which the first pie slice starts,using it you can rotate the entire pie chart
#counterclock - parameter that determines the direction in which the pie slices are drawn

plt.title("Split among classes")
plt.axis('equal')
plt.show()
```

# A4: Histograms

```python
a = [1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]

plt.hist(a)

plt.show()
```

```python
plt.hist(a, edgecolor='black')
plt.yticks([0,2,4,6,8,10])
#yticks([]) - it is used  to customize the placement and labels of the tick marks on the y-axis

plt.show()
```

Bins - They are used to divide the entire range of values into a set of intervals. Height of each bin represents the frequency or count of data points that fall within that interval.



```python
# How to see which numbers' frequencies are linked together?

x = np.arange(0,22)
plt.hist(a, edgecolor='black')
plt.xticks(x)

plt.show()

# Size of each bin = [(max value - min value) / (total bins)] = (22-1)/10 = 2.1
```

```python
x = np.arange(0,23)
plt.hist(a, edgecolor='black', bins = 21) # bins=21, sets width = [(22-1)/21] = 1
#bins = x, parameter allows us to explicitly set the desired number of bins for the histogram

plt.xticks(x)
plt.show()
```

```python
plt.hist(a, edgecolor='black', bins = [1,3,6,10,15,20])

#The bins are defined as [1, 3, 6, 10, 15, 20], which means the data will be divided into the following intervals: (1, 3], (3, 6], (6, 10], (10, 15], (15, 20]

plt.show()
```

```python
# axes scale
x=np.arange(25)
y=np.arange(10)
plt.hist(a, edgecolor='black', bins = [1,3,6,10,15,20])

plt.xticks(x+1)
plt.yticks(y+1)
#not mentioning these above 2 lines will cause graph to scale values of axes to be divided into intervals such as [0,5], [5,10], etc instead of consecutive terms

plt.axis([0,25,0,10]) #[x_min, x_max, y_min, y_max]
plt.show()
```

# A6: Stack Plot

```python
a = [1,2,3,4,5]
area = [2,1,6,2,10]

plt.title('Single Stack Plot')
plt.stackplot(a,area)
plt.show()
```

```python
a = [1,2,3,4,5]
area1 = [2,1,6,2,10]
area2 = [4,6,2,3,5]
area3 = [9,3,4,4,2]

labels=["area 1","area 2","area 3"]
colors=["purple","blue","green"]

plt.title('Multiple Stack Plot')
plt.stackplot(a,area1,area2,area3, labels=labels, colors=colors)

plt.legend(loc=2) #always use leged() with labels parameter
#loc=1, up-right
#loc=2, up-left
#loc=3, bottom-left
#loc=4, bottom-right
plt.show()
```
